```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv('kiva.csv')
```

```python
print(df.head(20))
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax.set_title("Kiva Loan Amounts by Country, Gender")
sns.set_palette('Pastel1')
sns.barplot(data=df, x='country', y='loan_amount', hue='gender', estimator=np.mean)
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
sns.set_style('darkgrid')
plt.show()
```

```python
plt.figure(figsize=(16, 10))
ax.set_title("Kiva Loan Amounts by Country, Distribution")
sns.set_palette('Pastel1')
sns.boxplot(data=df, x='country', y='loan_amount')
sns.set_style('darkgrid')
plt.show()
```

```python
plt.figure(figsize=(16, 10))
ax.set_title("Kiva Loan Amounts by Activty, Distribution")
sns.set_palette('Pastel1')
sns.violinplot(data=df, x='activity', y='loan_amount', hue='gender')
sns.set_style('darkgrid')
plt.show()
```
